```python
from __future__ import division, print_function
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('./augustus.xlsm', header=None,
                   names=['gene', 'hash', 'id', 'db', 'db_rec', 'type',
                          'n', 'm', 'k', 'l', 'date', 'what', 'descr'])
```

```python
df = df.drop(columns=['hash', 'id', 'db_rec', 'type', 'n', 'm', 'k',  'l', 'date', 'what']).dropna()
```

```python
what_to_find = 'HAD|Nicastrin'
search_result = df[df['descr'].str.contains(what_to_find)]
search_result
```

```python
search_result_no_duplicates = search_result.drop_duplicates(subset=['gene'])
search_result_no_duplicates
```
